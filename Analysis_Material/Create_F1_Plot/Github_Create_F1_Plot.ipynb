### Run the following code to import any libraries or packages required to run the rest of the script.

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Run the following code to generate a plot showing the F1-Scores for various confidence thresholds.

The user will need to adjust the base_Directory variable to match their directory structure.

```python
base_Directory = ""

f1_Scores = np.load(base_Directory + "Analysis/F1_Scores/F1_Scores.npy")
confidence_Values = np.linspace(0.5,0.99,len(f1_Scores))

fig, ax = plt.subplots()
plt.scatter(confidence_Values,f1_Scores, color = "blue", alpha = 0.5)
plt.xlabel("Confidence Threshold")
plt.ylabel("F1 Score")
plt.xlim(0.5,1)
plt.ylim(0,1)
plt.title("F1 Score Vs. Confidence Threshold")
plt.savefig(base_Directory + "Analysis/Outputs/F1_Scores.png")
plt.savefig(base_Directory + "Analysis/Outputs/F1_Scores.png")
plt.show()
```
