# Complete Analysis

 To complete the analysis, we first will **manually upload** our video to the cloned Github repository. To do this, on the left-hand side of the screen, open the Google Colab files, and navigate to the **motion-blur-microscopy -> Training_Material -> Extract_Video_Frames -> Video_Location** folder. 

You can then **right click on the Video_Location folder**, click upload, and then find the video you want to split on your computer. Once the video is uploaded, you can run the following block of code to split the video into frames.

```python
#@title Split Video Into Frames

video_Directory = "motion-blur-microscopy/Training_Material/Extract_Video_Frames/Video_Location/"
frame_Save_Location = "motion-blur-microscopy/Training_Material/Extract_Video_Frames/Video_Frames_Location/"

if '.ipynb_checkpoints' in os.listdir(video_Directory):
  os.rmdir(video_Directory + ".ipynb_checkpoints")

if '.ipynb_checkpoints' in os.listdir(frame_Save_Location):
  os.rmdir(frame_Save_Location + ".ipynb_checkpoints")

frame_Count = 0
for video in os.listdir(video_Directory):
  if video == "blank":
    os.remove(video_Directory + video)
    continue
  cap = cv2.VideoCapture(video_Directory + video)
  while(cap.isOpened()):
      frame_Count = frame_Count + 1
      ret, frame = cap.read()
      if ret == True:
          plt.imsave(frame_Save_Location + str(frame_Count) + ".png",frame)
      else:
          break

for frame in os.listdir(frame_Save_Location):
  if frame == "blank":
    os.remove(frame_Save_Location + video)
```

The video frames will be saved to the folder **motion-blur-microscopy -> Training_Material -> Extract_Video_Frames -> Video_Frames_Location**, where each frame in the video is labeled sequentially from 1 to the total number of frames in the video.
